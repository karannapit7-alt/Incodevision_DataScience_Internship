```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('global_green_energy_pulse_20260112.csv')
df.head()
```

```python
# Check how much columns and Rows
df.shape
```

```python
df
```

```python
df.describe()
df
```

```python
ny_df = df[df['city'] == 'New York']

plt.figure(figsize=(12,6))
plt.plot(ny_df['time'], ny_df['green_score'], label='Green Score')
plt.title('Green Energy Pulse Over Time (New York)')
plt.xlabel('Time')
plt.ylabel('Energy Score')
plt.legend()
plt.grid(True)
plt.show()
```
