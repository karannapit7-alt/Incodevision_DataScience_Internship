```python
import pandas as pd
```

```python
df = pd.read_csv('1_Intership_data.csv')
```

```python
df.head()
```

```python
# Check missing values
print(df.isnull().sum())
```

```python
# Age aur marks ko average (mean)se bharna
df['Age']= df['Age'].fillna(df['Age'].mean())
df['marks']= df['marks'].fillna(df['marks'].mean())
```

```python
df
```

```python
#1. columns ko strip kiye
df.columns = df.columns.str.strip()

# 2. Ab saari missing values
df['Name'] = df['Name'].fillna('Unknown Student')
df['Department'] = df['Department'].fillna('General')
df['city'] = df['city'].fillna('Not Disclosed')
df['Joining_Date'] = df['Joining_Date'].fillna('01-01-2025')
df['Phone_No'] = df['Phone_No'].fillna('0')

# 3. Numeric values (Age aur Marks) ko average.
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['marks'] = df['marks'].fillna(df['marks'].mean())

# 4. Final Cleaned file save
df.to_csv('Cleaned_Data_Final.csv', index=False)
```

```python
# Text ko uppercase main convert karna 
df['Department'] = df['Department'].str.upper()
df['city'] = df['city'].str.upper()
```

```python
# Duplicate rows delete karna
df = df.drop_duplicates()
```
